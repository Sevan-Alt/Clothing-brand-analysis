# Sales Performance & Customer Behaviour Analysis for an Online Clothing Retailer

### The company is an online fashion retail store specialising in casual unisex clothing. The brand offering its self-designed clothing line in 23 colours and 9 sizes. It operates solely online from the UK, shipping worldwide, with thoughts of opening a physical store.

```python
import mysql.connector
from dotenv import load_dotenv
import os
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
load_dotenv()

db = mysql.connector.connect(
    host=os.getenv("DB_HOST"),
    user=os.getenv("DB_USER"),
    password=os.getenv("DB_PASSWORD"),
    database=os.getenv("DB_NAME")
)
mycursor = db.cursor()
print(db)
```

# Understanding the data

###

- order_line_items - information on money received, returned and quantity of products
- order_marketing_details - transactions received from different marketign sources
- order_transactions - transaction information
